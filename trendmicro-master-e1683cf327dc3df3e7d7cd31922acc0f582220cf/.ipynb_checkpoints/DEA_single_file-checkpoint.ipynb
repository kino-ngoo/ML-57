```python
'''Basic package'''
import os

os.environ["CUDA_VISIBLE_DEVICES"] = "0"

import numpy as np
import pandas as pd
import tqdm
import time
tqdm.monitor_interval = 0

from multiprocessing import Pool
```

```python
# 分散式處理
def basic_analysis(filename):
    one_file = mainfile.loc[mainfile['FileID'] == filename]
    file_count = one_file.shape[0]
    user_count = len(one_file.CustomerID.unique())
    prod_count = len(one_file.ProductID.unique())
    time_diff =one_file.groupby(['CustomerID','QueryTS']).size().as_matrix()
    no_diff_count=0
    no_diff_count=time_diff.sum()-len(time_diff)
    append = pd.DataFrame([[filename,file_count,user_count,prod_count,no_diff_count]],columns=file_stat.columns)
    return append


def load_basic():
    # set up your pool
    pool = Pool(processes=10) # or whatever your hardware can support
    # have your pool map the file names to dataframes
    df_list = pool.map(basic_analysis, file_list)
    combined_df = pd.concat(df_list, ignore_index=True)
    return combined_df


def get_product_info(mainfile):
    ##product info
    #用法可參考：https://stackoverflow.com/questions/14529838/apply-multiple-functions-to-multiple-groupby-columns
    product = mainfile.groupby(['FileID','ProductID']).size().to_frame()
    product =product.reset_index()
    product.columns=['FileID','ProductID','times']
    f = {'times':['max','min','mean','std']}
    product_info=product.groupby('FileID').agg(f)
    product_info.columns=product_info.columns.get_level_values(1)
    product_info['std'] = product_info['std'].fillna(0)
    product_info.columns=['prod_open_max','prod_open_min','prod_open_mean','prod_open_std']
    return product_info

def time_info_analysis(filename):
    time_list=[]
    time_interval=[]

    samefile = mainfile[(mainfile.FileID==filename)]
    customer_list=group_list.loc[group_list.FileID==file_list[i]].CustomerID.values
    for j in np.arange(len(customer_list)):  
        
        samecustomer=samefile[samefile.CustomerID==customer_list[j]]
        time_list=samecustomer.QueryTS.values
        for k in np.arange(len(time_list)):
            if k+1<len(time_list):
                time_interval.append(time_list[k+1]-time_list[k])
    time_interval=np.array(time_interval)
    if len(time_interval)==0:
        time_interval_min=0
        time_interval_max=0
        time_interval_mean=0
        time_interval_std=0
    else:
        time_interval_min=time_interval.min()
        time_interval_max=time_interval.max()
        time_interval_mean=time_interval.mean()
        time_interval_std=time_interval.std()
    time_list=[]
    time_interval=[]
    append = pd.DataFrame([[filename,time_interval_min,time_interval_max,time_interval_mean,time_interval_std]],
                         columns=['FileID','interval_min','interval_max','interval_mean','interval_std'])
    return append


def load_time_info():

    # set up your pool
    pool = Pool(processes=10) # or whatever your hardware can support
    # have your pool map the file names to dataframes
    df_list_time = pool.map(time_info_analysis, file_list)
    combined_time = pd.concat(df_list_time, ignore_index=True)
    return combined_time
```

```python
data_dir='/data/examples/trend/data/query_log/'
dir_list = os.listdir(data_dir)
dir_list.sort()
```

```python
file_stat = pd.DataFrame(columns=['FileID','file_count', 'user_count','prod_count','no_diff_count'])

for i in tqdm.tqdm_notebook(np.arange(len(dir_list)), desc='file'):
    mainfile = pd.read_csv('/data/examples/trend/data/query_log/%s' %dir_list[i],
                        names=('FileID', 'CustomerID', 'QueryTS', 'ProductID'),
                        header=None,
                        dtype='object,object,int64,object')
    file_list = mainfile.FileID.unique().tolist()
    group_list =mainfile.groupby(['FileID','CustomerID']).size().to_frame().reset_index()
    product_info=get_product_info(mainfile)
    basic_info=load_basic()
    time_info=load_time_info()

    product_info['FileID']=product_info.index
    result = pd.merge(pd.merge(basic_info,product_info,how='inner',on='FileID'),time_info,how='inner',on='FileID')
    result.to_csv('temp/%s'%dir_list[i],index=False)
    print('finish file num: %s' %(i+1))
```
